```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
# these are our observations for the predictor - we will use BMI
x = np.array([1.5,2,2.5,3,3.5,4,4.5,5,5.5,6])
# these are the actual responses Y
y = np.array([10.35,12.3,13,14.0,16,17,18.2,20,20.7,22.5])
# We get the gradient (slope) and intercept for the line of best fit
gradient, intercept, r_value, p_value, std_err = stats.linregress(x,y)
mn=np.min(x)
mx=np.max(x)
# here an x1 array is created, with extra (points to increase the resolution?)
x1=np.linspace(mn,mx,500)
# a corresponding array of predicted response values y1 is generated, based
# on intercept and slope generated from the stats.linregress function, plus
y1=gradient*x1+intercept
# the sparser x, y arrays are plotted as a scatter plot
plt.plot(x,y,'ob')
# followed by the denser, higher resolution x1, y1 as a line
plt.plot(x1,y1,'-r')
plt.show()
print("intercept = ", intercept)
```

```python
# We will do the same for our BMI predictor, except in this case we will use the predictor BMI's 
# observations as the x axis values both for the scatter plot of observations, as well as for
# the line of best fit
# also, we are naming our slope "slope" as per documentation, and not "gradient" as per example,
# to keep naming consistent with documentation
# BMI data
x = ddata[:,2]
y = ddata[:,10]
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
y1=slope*x+intercept
```

```python
# There are a lot of potential improvements here, the graph could have x and y axes scaled to better
# reflect data. For now, we will print the basic markers and generate a graph, wrap the code in a function
# to use it in a for loop

def myplot(x, y, label):
    %matplotlib inline
    import numpy as np
    import matplotlib.pyplot as plt 
    from scipy import stats
    # assume a matplotlit.pyplot plt object has been defined
    slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    print("slope = ", slope)
    print("intercept = ", intercept)
    # r_value is the correlation coefficient
    print("r_value = ", r_value) 
    print("p_value = ", p_value)
    print("std_err = ", std_err)
    y1=slope*x+intercept
    fig, ax = plt.subplots()
    ax.plot(x,y,'ob')
    ax.plot(x,y1,'-r')

    ax.set(xlabel=label, ylabel='Response',
           title='Observed responses and line of best fit for ' + label + ' predictor')
    ax.grid()
    # TODO scale x and y, otherwise intercept might not look good
    fig.set_figheight(4)
    fig.set_figwidth(15)
    fig.savefig("images/" + label + "-LineOfBestFit.png")
    return plt # plt.show()
```

```python
plt = myplot(x, y, "BMI")
plt.show()
```

```python
# Ok. Have function, will plot
# reassign y just to remind us where the vector is coming from
y = ddata[:,10]
for j in range((ddata.shape[1] - 1)):
    x = ddata[:,j]
    #print(j)
    #print(dheaders[j])
    plt = myplot(x, y, dheaders[j])
    plt.show()
```
