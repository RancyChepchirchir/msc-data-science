### Import data

Our data is in hexadecimal format so for starters we will convert to comma separated integers, save csv files and work with csv thereafter

```python
######################################
### PROCESS LIGHTER FLAME LOG FILE ###
######################################

# vars
# expected record length
iRecLen = 12
# lists of values
iFA = []
iFB = []
iG= []

logfilename = "data/Test45.log"
file = open(logfilename, "r")
for line in file:
    line = line.rstrip()
    if(len(line) == iRecLen):
        # FlameA, FlameB and Guard detector analogue readings
        iFA.append(int(line[0:2], 16))
        iFB.append(int(line[2:4], 16))
        iG.append(int(line[4:6], 16))
        
# Lighter flame data lists converted to numpy arrays
iFAlfd = np.asarray(iFA)
iFBlfd = np.asarray(iFB)
iGlfd = np.asarray(iG) 

# Create index array for our x axis time series
iIdx = np.arange(1, iGlfd.size + 1, 1)
print("LIGHTER FLAME DATA")
print("FlameA array size = " + str(iFAlfd.size))
print("FlameB array size = " + str(iFBlfd.size))
print("Guard array size = " + str(iGlfd.size))
print("Index iIdx count = " + str(iIdx.size))
```

```python
# Now let's have a look at this data
# We know that analogue readings are acquired every 10ms so that gives us the base of our time series on the X axis
# We also know that data is represented as an unsigned byte ranging from 0 to 255 
# giving us the interval for our Y axis
# First we look at Flame A detector for our barbecue lighter data using pyplot from matplotlib
import matplotlib.pyplot as plt
plt.clf()
plt.plot(iIdx, iFAlfd, label='FlameA')
plt.xlabel('Time(ms)')
plt.ylabel('AD values')
plt.title("Flame lighter data")
plt.legend()
plt.show()
```

```python
# let's stretch out the x axis to better visualise data using figure figsize argument
plt.clf()
plt.figure(figsize=(18,8))
plt.plot(iIdx, iFAlfd, label='FlameA')
plt.xlabel('Time(ms)')
plt.ylabel('AD values')
plt.title("Flame lighter data")
plt.legend()
plt.show()
```

```python
# That basically suggests barbecue lighter was lit somewhere after the 1000ms mark up and extinguished 
# somewhere near the 2000ms mark.
# Now let's superimpose FlameB and Guard to further visualise our data
plt.clf()
plt.figure(figsize=(18,8))
plt.plot(iIdx, iFAlfd, label='FlameA')
plt.plot(iIdx, iFBlfd, label='FlameB')
plt.plot(iIdx, iGlfd, label='Guard')
plt.xlabel('Time(ms)')
plt.ylabel('AD values')
plt.title("Flame lighter data")
plt.legend()
plt.show()
```

```python
# We see that FlameA and FlameB are providing very similar readings with minor sections of no overlap (around 1000ms)
# Since the area of interest is between 1000ms and 2000ms, it makes sense to discard the rest of our data and focus
# on the area of interest, "zooming in" so to speak
# Todo do so, we slice the arrays between 950ms and 2100ms, so index counter 950 to 2100
iFAlfd_sliced = iFAlfd[950:2100:1]
iFBlfd_sliced = iFBlfd[950:2100:1]
iGlfd_sliced = iGlfd[950:2100:1]
iIdx_sliced = np.arange(1, iGlfd_sliced.size + 1, 1)
# Let's look at our new array sizes
print("SLICED LIGHTER FLAME DATA")
print("FlameA sliced array size = " + str(iFAlfd_sliced.size))
print("FlameB sliced array size = " + str(iFBlfd_sliced.size))
print("Guard sliced array size = " + str(iGlfd_sliced.size))
print("Index sliced iIdx count = " + str(iIdx_sliced.size))
```

```python
# Now we plot our sliced data
plt.clf()
plt.figure(figsize=(18,8))
plt.plot(iIdx_sliced, iFAlfd_sliced, label='FlameA sliced')
plt.plot(iIdx_sliced, iFBlfd_sliced, label='FlameB sliced')
plt.plot(iIdx_sliced, iGlfd_sliced, label='Guard sliced')
plt.xlabel('Time(ms)')
plt.ylabel('AD values')
plt.title("Sliced Flame lighter data")
plt.legend()
plt.show()
```

```python
# now we redo our fast fourier transform with the narrowed down data
start = 95
end = 201
t = iFArfd_sliced[start:end]
sp = np.fft.fft(np.sin(t))
freq = np.fft.fftfreq(t.shape[-1])
plt.figure(figsize=(18,8))
plt.plot(freq, sp.real)
plt.show()
```

```python
# our signal is obviously very noisy
# let's try to smooth it first using the Savistizky-Golay algorithm, which is basically a least squares distance function
import scipy.signal as sig
x = iFArfd_sliced[start:end]
t = sig.savgol_filter(x, 5, 2)
sp = np.fft.fft(np.sin(t))
freq = np.fft.fftfreq(t.shape[-1])
plt.figure(figsize=(18,8))
plt.plot(freq, sp.real)
plt.show()
```

```python
# fft looks not much different, maybe this approach is not helping us, lets try smoothing the original plots with the 
# savistzky golay algorithm
# first just one small section of a signal
plt.plot(x)
plt.show()
```

```python
# now the smoothed version
plt.plot(t)
plt.show()
```

```python
# ok, better, let's try using higher degree polynomials, remembering it cannot be greater than the window length (5)
t = sig.savgol_filter(x, 5, 4)
plt.plot(t)
plt.show()
```

```python
# ok, that did actually make it worse,
# let's step back and keep the polynomial order at two but increase the window length
t = sig.savgol_filter(x, 11, 2)
plt.plot(t)
plt.show()
```

```python
# that looks a lot smoother, except the actual spike is looking spikier, let's live with that for now and increase
# the window length again, remembering, window length must be an odd number
t = sig.savgol_filter(x, 17, 2)
plt.plot(t)
plt.show()
```

```python
# now that is a lot smoother, compared to the original data, let's replot just to remind us how noisy it was
t = iFArfd_sliced[start:end]
plt.plot(t)
plt.show()
```

```python
# one last attempt at increasing the polynomial order, this time to 3.
t = sig.savgol_filter(x, 17, 3)
plt.plot(t)
plt.show()
```

```python
# 3 order polynomial seems to work better than two, in the sense that it preserves the roundness of the spike better,
# keeping that characteristic of the signal, so let's go with that, 17 unit window length and 3 order polynomial

# Before we move on, let's try the fft one last time with our optimal window length and polyorder
x = iFArfd_sliced[start:end]
window_length = 17
polyorder = 3
t = sig.savgol_filter(x, window_length, polyorder)
sp = np.fft.fft(np.sin(t))
freq = np.fft.fftfreq(t.shape[-1])
plt.figure(figsize=(18,8))
plt.plot(freq, sp.real)
plt.show()
```

```python
# Admitedly, it's starting to look more like a sinusoidal wave, but really, it does not look usable so let's forget
# about fft for now

# Now we plot our filtered RF interference sliced data
window_length = 17
polyorder = 3
iFArfd_sliced_SG_filtered = sig.savgol_filter(iFArfd_sliced, window_length, polyorder)
iFBrfd_sliced_SG_filtered = sig.savgol_filter(iFBrfd_sliced, window_length, polyorder)
iGrfd_sliced_SG_filtered = sig.savgol_filter(iGrfd_sliced, window_length, polyorder)

plt.clf()
plt.figure(figsize=(18,8))
plt.plot(iIdx_sliced, iFArfd_sliced_SG_filtered, label='Filtered RF interference FlameA sliced')
plt.plot(iIdx_sliced, iFBrfd_sliced_SG_filtered, label='Filtered RF interference FlameB sliced')
plt.plot(iIdx_sliced, iGrfd_sliced_SG_filtered, label='Filtered RF interference Guard sliced')
plt.xlabel('Time(ms)')
plt.ylabel('AD values')
plt.title("RF interference Sliced data - filtered with the Savistzky-Golay algorithm")
plt.legend()
plt.show()
```

```python
# Now we apply the same filter to our flame data
window_length = 17
polyorder = 3
iFAlfd_sliced_SG_filtered = sig.savgol_filter(iFAlfd_sliced, window_length, polyorder)
iFBlfd_sliced_SG_filtered = sig.savgol_filter(iFBlfd_sliced, window_length, polyorder)
iGlfd_sliced_SG_filtered = sig.savgol_filter(iGlfd_sliced, window_length, polyorder)

plt.clf()
plt.figure(figsize=(18,8))
plt.plot(iIdx_sliced, iFAlfd_sliced_SG_filtered, label='Filtered FlameA sliced')
plt.plot(iIdx_sliced, iFBlfd_sliced_SG_filtered, label='Filtered FlameB sliced')
plt.plot(iIdx_sliced, iGlfd_sliced_SG_filtered, label='Filtered Guard sliced')
plt.xlabel('Time(ms)')
plt.ylabel('AD values')
plt.title("Filtered Sliced Flame lighter data")
plt.legend()
plt.show()
```
